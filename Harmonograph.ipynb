# Creating a Harmonograph with python3


```python
# Primero importamos la librerías que necesitamos

import numpy as np
import matplotlib.pyplot as plt
```

### Partimos de la ecuación 

$x(t) = Asin(tf + p)e^-dt$,

donde $f$ es la frecuencia, $p$ es la fase, $A$ es la amplitud, $d$ representa el amortiguamiento, $t$ es el tiempo.

Con las siguientes ecuaciones generamos la gráfica deseada,

$x(t) = A_1 sin(tf_1 + p_1)e^{-d_1t} + A_2 sin(tf_2 + p_2)e^{-d_2t}$

$y(t) = A_3 sin(tf_3 + p_3)e^{-d_3t} + A_4 sin(tf_4 + p_4)e^{-d_4t}$

```python
# Asignamos valores a los parámetros

n = 10000000
t = np.logspace(np.log10(10), np.log10(500),n)

# Asignamos valores para las constantes de las ecuaciones

A = [1, 1, 1.5, 1.5, 2, 2, 3]
d = [0.004, 0.001, 0.002, 0.0015, 0.1, 1, 2]
f = [3, 1 , 2, 2.5, 3, 10, 11]

x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

#Graficamos

plt.plot(x, y , 'k', linewidth=1, color = 'aqua')
plt.axis('off')
plt.show()
```
